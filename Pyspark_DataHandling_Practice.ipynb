```python
#drop null values 
```

```python
df_pyspark=spark.read.option('header','true').csv('test.csv',inferSchema=True)
```

```python
df_pyspark.show()
```

```python
df_pyspark.na.drop().show()
```

```python
df_pyspark.na.drop(how='all').show()
```

```python
df_pyspark.na.drop(how='any',thresh=2).show() #atleast 2 non null values 
```

```python
df_pyspark.na.drop(how='any',subset=['exp']).show()
```

```python
df_pyspark.na.fill('Missing Values','exp').show()
```

```python
from pyspark.ml.feature import Imputer
imputer = Imputer(inputCols=['Age','exp'],
                 outputCols=["{}_imputed".format(c) for c in ['Age','exp']]).setStrategy("mean")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```

```python
from pyspark.ml.feature import Imputer
imputer = Imputer(inputCols=['Age','exp'],
                 outputCols=["{}_imputed".format(c) for c in ['Age','exp']]).setStrategy("median")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```

```python
from pyspark.ml.feature import Imputer
imputer = Imputer(inputCols=['Age','exp'],
                 outputCols=["{}_imputed".format(c) for c in ['Age','exp']]).setStrategy("mode")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```

```python
from pyspark.sql import SparkSession
spark =SparkSession.builder.appName('dataframe').getOrCreate()
```

```python
df_spark =spark.read.csv('test.csv',header=True,inferSchema=True)
```

```python
df_spark.show()
```

```python
df_spark.filter("exp <=5").show()
```

```python
df_spark.filter("exp <=5").select(['Name',"Age"]).show()
```

```python
df_spark.filter((df_spark['exp'] <=5) & (df_spark['Age'] >25)).show()
```

```python
df_spark.filter(~(df_spark['exp'] <=5)).show() #Not operation
```

```python
#Groupby Aggregate functions
```

```python
df_pyspark=spark.read.option('header','true').csv('test.csv',inferSchema=True)
```

```python
df_pyspark.printSchema() 
```

```python
df_pyspark.groupBy('Name').sum().show()
```

```python
df_pyspark.groupBy('Dept').sum().show() #deptwise
```

```python
df_pyspark.groupBy('Dept').mean().show() #deptwise
```
